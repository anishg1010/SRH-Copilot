# Teaching Agent — Agentic Demo (local, free)

Runs on the **local open-source LLM (Ollama)** with **Chroma-grounded tools**. The agent
picks tools (search knowledge, draft objectives, design assessment...), runs them, and
composes a grounded answer. No paid API.

**Prereqs:** `ollama serve` + `ollama pull qwen2.5:3b-instruct`, and
`python -m copilot.rag.ingest teaching ./data/teaching --reset`
